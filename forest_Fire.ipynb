# Algerian Forest Fire Data

The Algerian Forest Fires Dataset contains information about forest fires that occurred in two regions of Algeria: Bejaia and Sidi Bel-Abbes, from June to September 2012. This dataset is valuable for understanding the patterns and causes of forest fires in Algeria and can be used to predict future forest fires.

Dataset Features

The dataset includes the following features:

Date (DD/MM/YYYY): Day, month (June to September), year (2012).

Weather data observations: Temp: Temperature at noon (maximum temperature) in Celsius degrees (22 to 42). RH: Relative Humidity in % (21 to 90). Ws: Wind speed in km/h (6 to 29). Rain: Total rainfall in mm (0 to 16.8).

FWI Components: Fine Fuel Moisture Code (FFMC): Index from the FWI system (28.6 to 92.5). Duff Moisture Code (DMC): Index from the FWI system (1.1 to 65.9). Drought Code (DC): Index from the FWI system (7 to 220.4). Initial Spread Index (ISI): Index from the FWI system (0 to 18.5). Buildup Index (BUI): Index from the FWI system (1.1 to 68). Fire Weather Index (FWI): Index (0 to 31.1).

Classes: Two classes, namely: fire and not fire.

## Region Feature Creation

The dataset contains observations from two regions:
- Bejaia
- Sidi Bel-Abbes

A new column "Region" is created to distinguish between these regions:
- 0 → Bejaia
- 1 → Sidi Bel-Abbes

The dataset is split based on index position.

```python
ff.loc[:122,"Region"]=0 #Bejaia
ff.loc[123:,"Region"]=1 #Sidi Bel-Abbes
```

```python
ff
```

```python
ff.info()
```

```python
ff[['Region']]=ff[['Region']].astype(int)
```

```python
ff.head()
```

```python
ff[ff.isnull().any(axis=1)]
```

```python
ff=ff.dropna().reset_index(drop=True)
```

```python
ff.isnull().sum()
```

```python
ff.iloc[122]
```

The row at index 122 contained invalid or inconsistent data (column name) and is removed to maintain data quality.

```python
ff=ff.drop(122).reset_index(drop=True)
```

```python
ff.iloc[122]
```

```python
#fixing spaces in column names
ff.columns=ff.columns.str.strip()
ff.columns
```

```python
ff.info()
```

```python
ff.rename(columns={'Classes':'Fire_Status'}, inplace=True)
```

## Data Type Conversion

Several columns were stored as object (string) types. These were converted into numerical format using pd.to_numeric() to enable proper analysis and modeling.

```python
cols=['day','month','year','Temperature','RH','Ws','Rain','FFMC','DMC','DC','ISI','BUI','FWI']
ff[cols]=ff[cols].apply(pd.to_numeric,errors='coerce')	
```
